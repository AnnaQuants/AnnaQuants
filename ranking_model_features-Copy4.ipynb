```python
import yfinance as yf
import pandas as pd

# Function to download historical data for a single ticker with start and end dates
def download_data(ticker, start, end):
    data = yf.download(ticker, start=start, end=end)
    if 'dst_error_hours' in data.columns:
        # Correct the timedelta handling
        data.index += pd.to_timedelta(data['dst_error_hours'], unit='h')
        data.drop(columns='dst_error_hours', inplace=True)
    data['ticker'] = ticker
    return data

# Define the start and end dates for the data download
start_date = '2019-07-20'  # Example start date
end_date = '2024-07-20'    # Example end date

# Assuming your dataframe "df" has a column named 'ticker' with the list of tickers
tickers = df['ticker'].tolist()

# Download data for all tickers and concatenate into a single DataFrame
data_frames = [download_data(ticker, start_date, end_date) for ticker in tickers]
all_data = pd.concat(data_frames)
```

```python
import pandas as pd
import numpy as np

def ewma_half_life(data, span):
    alpha = 1 - np.exp(np.log(0.5) / span)
    return data.ewm(alpha=alpha, adjust=False).mean()

# Step 1: Calculate the EWMA for different time scales
short_scales = [8, 16, 32]
long_scales = [24, 48, 96]

for s in short_scales:
    all_data[f'EWMA_{s}'] = all_data.groupby('ticker')['Close'].transform(lambda x: ewma_half_life(x, s))

for l in long_scales:
    all_data[f'EWMA_{l}'] = all_data.groupby('ticker')['Close'].transform(lambda x: ewma_half_life(x, l))

# Step 2: Calculate the 63-day rolling standard deviation
all_data['Rolling_Std_63'] = all_data.groupby('ticker')['Close'].transform(lambda x: x.rolling(window=63).std())

# Step 3: Compute the volatility-scaled MACD
for s in short_scales:
    for l in long_scales:
        if l > s:
            all_data[f'MACD_{s}_{l}'] = (all_data[f'EWMA_{s}'] - all_data[f'EWMA_{l}']) / all_data['Rolling_Std_63']

# Step 4: Combine the volatility-scaled MACDs into a composite signal
macd_columns = [f'MACD_{s}_{l}' for s in short_scales for l in long_scales if l > s]
all_data['Composite_MACD'] = all_data[macd_columns].sum(axis=1)

# Handle NaNs using forward fill followed by backward fill
all_data.ffill(inplace=True)
all_data.bfill(inplace=True)

# Display the first few rows to check the new columns
print(all_data.head())
```

```python
import xgboost as xgb
import pandas as pd

# Step 1: Prepare the Data
# Assuming 'all_data' is your multi-index DataFrame with Date and ticker

# Flatten the multi-index for compatibility with XGBoost
all_data = all_data.reset_index()

# Define features and target
features = ['3Day_Ret_Class', 'Weekly_Ret_Class', 'RSI', 'ADX', 'Weekly_Return_Lag_1', 'Weekly_Return_Lag_2',
           'Weekly_Return_Lag_3', 'Weekly_Return_Lag_4', 'Weekly_Return_Lag_5', 'Weekly_Volatility', 
            'Weekly_RVOL', 'Weekly_ROC', 'MACD_8_24', 'MACD_16_24', 'MACD_16_96', 'Weekly_Return_Lag_1_Rank'
           ]

# Rank the weekly returns and rescale the ranks to fit within the range 0 to 31
all_data['Weekly_Return_Rank'] = all_data.groupby('Date')['Weekly_Return'].rank(method='first', ascending=False)
all_data['Weekly_Return_Rank'] = ((all_data['Weekly_Return_Rank'] - 1) * (31 / (all_data['Weekly_Return_Rank'].max() - 1))).astype(int)

# Use the ranks as the target
target = 'Weekly_Return_Rank'

# Create query ID column based on Date
all_data['qid'] = all_data['Date'].factorize()[0]

# Split the data into training and testing sets
train_data = all_data[all_data['Date'] < '2023-06-01']
test_data = all_data[all_data['Date'] >= '2023-06-01']

# Create DMatrix for XGBoost with query ID
dtrain = xgb.DMatrix(train_data[features], label=train_data[target])
dtrain.set_group(train_data.groupby('qid').size().to_list())

dtest = xgb.DMatrix(test_data[features], label=test_data[target])
dtest.set_group(test_data.groupby('qid').size().to_list())

# Step 2: Train the LambdaMART Model
params = {
    'objective': 'rank:pairwise',  # Pairwise ranking objective
    'eval_metric': 'ndcg',         # NDCG metric for evaluation
    'eta': 0.1,
    'gamma': 1.0,
    'min_child_weight': 0.1,
    'max_depth': 6,
    'nthread': 4,
    'tree_method': 'hist',
    'device': 'cuda'  # Use GPU acceleration if available
}

num_rounds = 100  # Number of boosting rounds
model = xgb.train(params, dtrain, num_rounds, evals=[(dtest, 'eval')], early_stopping_rounds=10)

# Step 3: Evaluate and Predict
predicted_scores = model.predict(dtest)

# Convert predicted scores to ranks (ascending order of scores)
test_data['Predicted_Score'] = predicted_scores
test_data['Predicted_Rank'] = test_data.groupby('Date')['Predicted_Score'].rank(method='first', ascending=True)

# Restore multi-index
test_data = test_data.set_index(['Date', 'ticker'])

# Select a specific date for inspection
specific_date = '2024-07-12'
data_for_date = test_data.xs(specific_date, level='Date')

# Display the top 10 and bottom 10
top_10 = data_for_date.nsmallest(10, 'Predicted_Rank')
bottom_10 = data_for_date.nlargest(10, 'Predicted_Rank')

print("Top 10 Tickers by Predicted Rank on", specific_date)
print(top_10[['Weekly_Return', 'Weekly_Return_Rank', 'Predicted_Score', 'Predicted_Rank']])

print("\nBottom 10 Tickers by Predicted Rank on", specific_date)
print(bottom_10[['Weekly_Return', 'Weekly_Return_Rank', 'Predicted_Score', 'Predicted_Rank']])
```

```python
import yfinance as yf

# Function to calculate weekly return
def calculate_weekly_return(data):
    return (data[-1] - data[0]) / data[0]

# Function to calculate the new portfolio value
def calculate_portfolio_value(initial_investment, returns, short=False):
    if short:
        returns = -returns  # Invert the returns for short positions
    return initial_investment * (1 + returns)

# List of top and bottom tickers on 2024-06-21
top_tickers = ['AMT', 'AOS', 'ARE', 'ARGX', 'AWK', 'BEP', 'BIDU', 'FND', 'BIP', 'BLD']
bottom_tickers = ['HUBS', 'CAVA', 'WING', 'NTNX', 'CMG', 'MNDY', 'FTAI', 'META', 'NFLX', 'NOW']

# Fetch data from yfinance for the given date range
start_date = '2024-07-14'
end_date = '2024-07-20'

tickers = top_tickers + bottom_tickers
data = yf.download(tickers, start=start_date, end=end_date)

# Calculate weekly returns
weekly_returns = {}
for ticker in tickers:
    ticker_data = data['Adj Close'][ticker]
    weekly_returns[ticker] = calculate_weekly_return(ticker_data)

# Initialize portfolio value
initial_portfolio_value = 40000  # Total initial portfolio value
individual_investment = initial_portfolio_value / len(tickers)

# Calculate new portfolio value
new_portfolio_value = 0
for ticker in top_tickers:
    new_value = calculate_portfolio_value(individual_investment, weekly_returns[ticker], short=False)
    new_portfolio_value += new_value

for ticker in bottom_tickers:
    new_value = calculate_portfolio_value(individual_investment, weekly_returns[ticker], short=True)
    new_portfolio_value += new_value

# Display the actual weekly returns
print("Actual Weekly Returns from 2024-07-14 to 2024-07-20")

print("\nTop 10 Predicted Tickers on 2024-07-20:")
for ticker in top_tickers:
    print(f"{ticker}: {weekly_returns[ticker]:.6f}")

print("\nBottom 10 Predicted Tickers on 2024-07-20:")
for ticker in bottom_tickers:
    print(f"{ticker}: {weekly_returns[ticker]:.6f}")

# Display the portfolio value
print(f"\nInitial Portfolio Value: ${initial_portfolio_value}")
print(f"New Portfolio Value: ${new_portfolio_value:.2f}")
print(f"Total Return: {((new_portfolio_value - initial_portfolio_value) / initial_portfolio_value) * 100:.2f}%")
```

```python
# Load the CSV file, preserving the Date and ticker multi-index
all_data = pd.read_csv('mega_historical_data4.csv', index_col=['Date', 'ticker'], parse_dates=['Date'])
```

```python
import xgboost as xgb
import pandas as pd
import numpy as np

# Step 1: Prepare the Data
# Assuming 'all_data' is your multi-index DataFrame with Date and ticker

# Flatten the multi-index for compatibility with XGBoost
all_data = all_data.reset_index()

# Define features and target
features = ['3Day_Ret_Class', 'Weekly_Ret_Class', 'RSI', 'ADX', 'Weekly_Return_Lag_1', 'Weekly_Return_Lag_2',
            'Weekly_Return_Lag_3', 'Weekly_Return_Lag_4', 'Weekly_Return_Lag_8', 'Weekly_Volatility', 
            'Weekly_RVOL', 'Weekly_ROC', 'MACD_8_24', 'MACD_16_24', 'MACD_16_96', 'Weekly_Return_Lag_1_Rank']  
target = 'Weekly_Return_Rank'  # Relevance is now Weekly_Return_Rank

# Ensure the relevance degrees are within the acceptable range
all_data[target] = all_data[target].clip(upper=31)

# Invert the relevance labels to ensure higher ranks are treated as more relevant
max_rank = all_data[target].max()
all_data[target] = max_rank - all_data[target]

# Split the data into training and testing sets
train_data = all_data[all_data['Date'] < '2022-07-12']
test_data = all_data[all_data['Date'] >= '2022-07-12']

# Define qids (query ids) for the groups based on the Date
train_qids = train_data['Date'].factorize()[0]
test_qids = test_data['Date'].factorize()[0]

# Create group lists for training and testing data
train_group = train_data['Date'].value_counts().sort_index().values
test_group = test_data['Date'].value_counts().sort_index().values

# Define the DMatrix for XGBoost
dtrain = xgb.DMatrix(train_data[features], label=train_data[target])
dtrain.set_group(train_group)
dtest = xgb.DMatrix(test_data[features], label=test_data[target])
dtest.set_group(test_group)

# Step 2: Train the XGBRanker Model
ranker = xgb.XGBRanker(
    objective='rank:pairwise',  # Pairwise ranking objective
    eval_metric='ndcg',         # NDCG metric for evaluation
    eta=0.1,
    gamma=1.0,
    min_child_weight=0.1,
    max_depth=6,
    nthread=4,
    tree_method='hist',
    early_stopping_rounds=10,
    device='cuda'  # Use GPU acceleration if available
)

# Train the model
ranker.fit(train_data[features], train_data[target], group=train_group, eval_set=[(test_data[features], test_data[target])], eval_group=[test_group])

# Step 3: Evaluate and Predict
predicted_scores = ranker.predict(test_data[features])

# Convert predicted scores to ranks
test_data['Predicted_Score'] = predicted_scores
test_data['Predicted_Rank'] = test_data.groupby('Date')['Predicted_Score'].rank(method='first', ascending=False)

# Restore multi-index
test_data = test_data.set_index(['Date', 'ticker'])

# Select a specific date for inspection
specific_date = '2024-07-19'
data_for_date = test_data.xs(specific_date, level='Date')

# Display the top 10 and bottom 10
top_10 = data_for_date.nsmallest(10, 'Predicted_Rank')
bottom_10 = data_for_date.nlargest(10, 'Predicted_Rank')

print("Top 10 Tickers by Predicted Rank on", specific_date)
print(top_10[['Weekly_Return', 'Weekly_Return_Rank', 'Predicted_Score', 'Predicted_Rank']])

print("\nBottom 10 Tickers by Predicted Rank on", specific_date)
print(bottom_10[['Weekly_Return', 'Weekly_Return_Rank', 'Predicted_Score', 'Predicted_Rank']])
```

```python
import yfinance as yf

# Function to calculate weekly return
def calculate_weekly_return(data):
    return (data[-1] - data[0]) / data[0]

# Function to calculate the new portfolio value
def calculate_portfolio_value(initial_investment, returns, short=False):
    if short:
        returns = -returns  # Invert the returns for short positions
    return initial_investment * (1 + returns)

# List of top and bottom tickers on 2024-06-21
top_tickers = ['SRPT', 'ADBE', 'SIRI', 'RYAN', 'TOST', 'GILD', 'UTHR', 'ACN', 'CHWY', 'ORLY']
bottom_tickers = ['YUMC', 'UWMC', 'UHAL', 'UAL', 'TM', 'TAK', 'SYM', 'STM', 'STLA', 'SQM']

# Fetch data from yfinance for the given date range
start_date = '2024-06-23'
end_date = '2024-06-29'

tickers = top_tickers + bottom_tickers
data = yf.download(tickers, start=start_date, end=end_date)

# Calculate weekly returns
weekly_returns = {}
for ticker in tickers:
    ticker_data = data['Adj Close'][ticker]
    weekly_returns[ticker] = calculate_weekly_return(ticker_data)

# Initialize portfolio value
initial_portfolio_value = 40000  # Total initial portfolio value
individual_investment = initial_portfolio_value / len(tickers)

# Calculate new portfolio value
new_portfolio_value = 0
for ticker in top_tickers:
    new_value = calculate_portfolio_value(individual_investment, weekly_returns[ticker], short=False)
    new_portfolio_value += new_value

for ticker in bottom_tickers:
    new_value = calculate_portfolio_value(individual_investment, weekly_returns[ticker], short=True)
    new_portfolio_value += new_value

# Display the actual weekly returns
print("Actual Weekly Returns from 2024-07-07 to 2024-07-12")

print("\nTop 10 Predicted Tickers on 2024-07-12:")
for ticker in top_tickers:
    print(f"{ticker}: {weekly_returns[ticker]:.6f}")

print("\nBottom 10 Predicted Tickers on 2024-07-12:")
for ticker in bottom_tickers:
    print(f"{ticker}: {weekly_returns[ticker]:.6f}")

# Display the portfolio value
print(f"\nInitial Portfolio Value: ${initial_portfolio_value}")
print(f"New Portfolio Value: ${new_portfolio_value:.2f}")
print(f"Total Return: {((new_portfolio_value - initial_portfolio_value) / initial_portfolio_value) * 100:.2f}%")

```
